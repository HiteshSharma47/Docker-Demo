```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc, sum, avg
```

```python
spark = SparkSession.builder.appName("DemoSpark").getOrCreate()
```

```python
data = [
    (1,"Hitesh","Engineer",3000,20),
    (2,"Sumit","Doctor",4000,30),
    (3,"Aman","Sales",5000,50),
    (4,"Amit","Commerce",6000,40),
]
```

```python
columns = ["ID", "Name", "Department", "Salary", "Age"]
```

```python
df = spark.createDataFrame(data, columns)
```

```python
print(df)
```

```python
df.show()
```

```python
df_col = df.select(["Name","Department","Salary"])
```

```python
df_col.show()
```

```python
df_filter = df.filter(col("Salary") > 4000)
```

```python
df_filter.show()
```

```python
for x in df_filter.collect():
    print(x)
```

```python
df_with_col = df.withColumn("Bonus", col("Salary") + col("Salary") * 0.10)
```

```python
df_with_col.show()
```

```python
df.write.mode("overwrite").csv("output/employeeTable", header=True)
```

```python
df.toPandas().to_csv(
    "output/employeesTableNew.csv",
    index=False
)
```

```python
df = spark.read \
.option("header",True) \
.csv("output/employeeTable")
```

```python
df.show()
```

```python
df.createOrReplaceTempView("employees") #Temp Name Of the spark.read file for sql
```

```python
result = spark.sql("""SELECT * FROM employees WHERE Salary > 4000""").show()
```
