```python
# UDF

data = [
    (1,"Hitesh", 23),
    (1,"Hanuman", 17),
    (1,"Aman", 12),
    (1,"Nikhil", 21),
    (1,"Raman", 18)
]

columns = ["ID", "Name", "Age"]

df = spark.createDataFrame(data,columns)
```

```python
df.show()
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def can_drive(age):
    if age > 18:
        return "Can Drive"
    elif age > 12 and age < 17:
        return "Can Drive License"
    else:
        return "Can't Drive"
```

```python
can_drive_udf = udf(can_drive,StringType())
```

```python
df_new = df.withColumn('Status', can_drive_udf('age'))
```

```python
df_new.show()
```
